```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Cargar el archivo CSV
file_path = ('Anexo 5Mall_Customers.csv')
data = pd.read_csv(file_path)

# Configurar el estilo de los grficos
sns.set(style="whitegrid")

# Grficos de distribucin para variables numricas
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Distribucin de la edad
sns.histplot(data['Age'], bins=15, kde=True, ax=axes[0], color="skyblue")
axes[0].set_title('Distribucin de la Edad')

# Distribucin del ingreso anual
sns.histplot(data['Annual Income (k$)'], bins=15, kde=True, ax=axes[1], color="orange")
axes[1].set_title('Distribucin del Ingreso Anual')

# Distribucin del puntaje de gasto
sns.histplot(data['Spending Score (1-100)'], bins=15, kde=True, ax=axes[2], color="green")
axes[2].set_title('Distribucin del Puntaje de Gasto')

plt.tight_layout()
plt.show()

# Grficos de relacin entre pares de variables
sns.pairplot(data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']])
plt.show()
```
